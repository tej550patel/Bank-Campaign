```python
bank_df = pd.DataFrame.copy(df) 
```

```python
bank_df.head() 
```

```python
#creating a list of categorical columns
colname = [] 
for x in bank_df.columns:
    if bank_df[x].dtype == 'object':
        colname.append(x) 
colname
```

```python
bank_df['job_description'].value_counts() 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() 

for x in colname:
    bank_df[x] = le.fit_transform(bank_df[x]) 
```

```python
bank_df.head() 
```

```python
X = bank_df[['age_in_years', 'job_description', 'marital_status',
       'education_details', 'has_default', 'balance_in_account',
       'housing_status', 'previous_loan', 'phone_type', 'date',
       'month_of_year', 'call_duration', 'campaign_contacts', 'days_passed',
       'previous_contact', 'poutcome_of_campaign']] 

y = bank_df['outcome']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(X) 
X = scaler.transform(X) 
print(X) 
```

```python
y = y.astype(int) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10) 
```

```python
#SMOTE
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2) 
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel()) 
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression() 
lr.fit(X_train_res, y_train_res.ravel()) 
```

```python
pred = lr.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, pred)) 
print(classification_report(y_test, pred)) 
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier() 
dtree.fit(X_train_res, y_train_res.ravel()) 
```

```python
tree_pred = dtree.predict(X_test) 
```

```python
print(confusion_matrix(y_test, tree_pred)) 
print(classification_report(y_test, tree_pred)) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier() 
rf.fit(X_train_res, y_train_res.ravel()) 
```

```python
rf_pred = rf.predict(X_test) 
```

```python
print(confusion_matrix(y_test, rf_pred)) 
print(classification_report(y_test, rf_pred)) 
```
